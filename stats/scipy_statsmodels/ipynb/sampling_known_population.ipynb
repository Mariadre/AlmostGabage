```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

%matplotlib inline
```

```python
# 以下のような一変量（魚の体長）データを考える
five_fish = np.array([2, 3, 4, 5, 6])
five_fish
```

```python
# ランダムサンプリング（重複なしで三つ）
np.random.choice(five_fish, size=3, replace=False)
```

```python
# 標本平均
sp.mean(np.random.choice(five_fish, size=3, replace=False))
```

```python
# 母集団として 100000尾の魚の体長データを考える
np.random.seed(8)
many_fish = pd.Series(np.random.normal(4, 0.8, 100000), name='length')
many_fish.head()
```

```python
# 母平均
print('mean: {:.2f}'.format(sp.mean(many_fish)))
# 母分散
print('std:  {:.2f}'.format(sp.var(many_fish, ddof=0)))
```

```python
# 10尾の魚をサンプリング
ten_sample_fish = np.random.choice(many_fish, size=10, replace=False)

# 標本平均
print('mean: {:.2f}'.format(sp.mean(ten_sample_fish)))
# 標本分散
print('std:  {:.2f}'.format(sp.var(ten_sample_fish, ddof=0)))
```

```python
# ヒストグラム
import warnings
warnings.filterwarnings('ignore')

sns.distplot(many_fish, kde=False)

# ヒストグラムの様子から、母集団を「平均 4, 分散 0.64」の正規分布と仮定してみる
```

```python
# 「平均 4, 分散 0.64」の正規分布を図示
x = np.arange(1, 7.1, 0.1)
plt.plot(x, stats.norm.pdf(x=x, loc=4, scale=0.8))
```

```python
# 母集団のヒストグラムと重ねてみる
sns.distplot(many_fish, kde=False, norm_hist=True)
plt.plot(x, stats.norm.pdf(x=x, loc=4, scale=0.8), color='blue', alpha=0.2)
```
