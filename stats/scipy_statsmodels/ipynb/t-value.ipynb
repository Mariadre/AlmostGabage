```python
import numpy as np
import pandas as pd
import scipy as sp
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

import warnings
warnings.filterwarnings('ignore')

%matplotlib inline
```

```python
# 母集団として平均 4, 標準偏差 0.8 の正規分布を考える
np.random.seed(1)
norm_dist = stats.norm(loc=4, scale=0.8)

# t値を求める
t_value = np.zeros(10000)
for i in range(10000):
    sample_values = norm_dist.rvs(size=10)
    sample_mu = sp.mean(sample_values)
    sample_se = sp.std(sample_values, ddof=1) / sp.sqrt(len(sample_values))
    t_value[i] = (sample_mu - 4) / sample_se
```

```python
# t値のヒストグラム
sns.distplot(t_value)
# 正規分布のヒストグラム
x = np.arange(-8, 8.1, 0.1)
plt.plot(x, stats.norm.pdf(x), linestyle=':')
```

```python
# t分布
plt.plot(x, stats.t.pdf(x, df=9))

# 正規分布
plt.plot(x, stats.norm.pdf(x), linestyle=':')
```
