```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

```python
# matplotlib のみでのグラフ描画
x = np.arange(0, 10)
y = np.array([2, 3, 4, 3, 5, 4, 6, 7, 4, 8])

plt.plot(x, y)
plt.title('lineplot sample')
plt.xlabel('x')
plt.ylabel('y')
```

```python
import seaborn as sns
sns.set()
```

```python
# jupyter lab 上で warning がうるさいので黙らせる
import warnings
warnings.filterwarnings('ignore')

# matplotlib のみでのグラフ描画
x = np.arange(0, 10)
y = np.array([2, 3, 4, 3, 5, 4, 6, 7, 4, 8])

plt.plot(x, y)
plt.title('lineplot sample')
plt.xlabel('x')
plt.ylabel('y')
```

```python
# 実数値で見たい場合は kde=False を指定
# 自分で階級を分けたい場合は bins=5 などと指定
# どちらも指定しない場合、カーネル密度推定による滑らかなヒストグラムになる
sns.distplot(y)
```

```python
cov_data = pd.DataFrame(
    {'x': [18.5, 18.7, 19.1, 19.7, 21.5, 21.7, 21.8, 22.0, 23.4, 23.8],
     'y': [34, 39, 41, 38, 45, 41, 52, 44, 44, 49]}
)
```

```python
# kind='reg' とすると回帰直線とその信頼区間も表示可能
sns.jointplot(x='x', y='y', data=cov_data)
```
