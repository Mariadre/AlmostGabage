```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from plotly import tools
from plotly.graph_objs import Bar, Figure, Layout, Scatter
from plotly.offline import init_notebook_mode, iplot

init_notebook_mode(connected=True)
%matplotlib inline
```

```python
df = pd.read_csv('../data/airpassengers.csv', index_col='Year')
date_index = pd.date_range('1949-01-01', freq='M', periods=144)

airpassengers = pd.DataFrame({
    'n_passengers': df.values.ravel()
})
airpassengers.index = date_index
airpassengers['n_passengers'] = airpassengers['n_passengers'].astype(np.float64)
airpassengers.head()
```

```python
data = [Scatter(x=airpassengers.index, y=airpassengers.n_passengers, name='n_passengers')]
fig = Figure(data=data, layout=Layout(title='旅客数'))
iplot(fig)
```
